```python
# check whether the number is palindrome is not 
palim = int(input("enter the number ")) 
check_palim = palim
new_num= 0 
while(palim!= 0 ): 
    last_num = palim % 10 
    new_num = new_num*10+ last_num
    palim //=10 
if (check_palim == new_num): 
    print("number is palimbome ")
else: 
    print("number is not palimbome ")

    
```
