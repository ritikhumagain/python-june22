```python
password = input("Enter the password: ")
password_len = len(password)

if password_len < 8:
    print("Password must be at least 8 characters")
else:
    has_digit = False
    has_upper = False
    has_special = False

    for i in password:
        if i.isdigit():
            has_digit = True
        elif i.isupper():
            has_upper = True
        elif not i.isalnum():
            has_special = True

    if has_digit and has_upper and has_special:
        print("Strong password")
    else:
        print("Weak password")
```
