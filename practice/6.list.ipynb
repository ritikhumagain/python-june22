```python
tasks = []

while True:
    print("\n--- TO-DO LIST ---")
    print("1. Add task")
    print("2. View tasks")
    print("3. Complete task")
    print("4. Quit")

    choice = input("\nEnter choice: ")

    if choice == "1":
        task = input("Enter task: ")
        tasks.append(task)
        print(f"✅ '{task}' added!")

    elif choice == "2":
        if len(tasks) == 0:
            print("No tasks yet!")
        else:
            print("\nYour Tasks:")
            for index, task in enumerate(tasks):
                print(f"{index + 1}. {task}")

    elif choice == "3":
        if len(tasks) == 0:
            print("No tasks to complete!")
        else:
            for index, task in enumerate(tasks):
                print(f"{index + 1}. {task}")
            num = int(input("Enter task number to complete: ")) - 1
            if 0 <= num < len(tasks):
                done = tasks.pop(num)
                print(f"✅ '{done}' completed and removed!")
            else:
                print("Invalid number!")

    elif choice == "4":
        print("Goodbye!")
        break

    else:
        print("Invalid choice!")
```
