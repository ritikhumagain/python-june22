```python
# Remove duplicates from a list (hint: you'll appreciate this problem more after)
name = list()
number = int(input("number of item you want to add "))
for i in range(number):
    name_user= input("enter the name ")
    name.append(name_user)
print(name)
new = list()
for j in name:
    if j not in new:
        new.append(j)
print(new)
```

```python
# Given a list of numbers, create a new list containing only the even numbers.
number= list()
total_number = int(input("Enter the total number "))
for i in range(total_number):
    number_user = int(input("enter the total number "))
    number.append(number_user)
print(f"the number you enter is : {number}")
even_number = list()
odd_number = list()
for j in number :
    if j%2 == 0 :
        if j  not in even_number:
            even_number.append(j)
    else:
        odd_number.append(j)
print("even Number is :",even_number)
print("Odd Number is :",odd_number)
```

```python
# . Given a list of words, create a new list with only words longer than 4
# characters (use a list comprehension)
name = list()
number = int(input("enter the no of itteration"))
for i in range(number):
    user_name = input("enter your name ")
    if len(user_name)<=4:
        name.append(user_name)
print(name)
```
